```python
from nltk import word_tokenize, download, pos_tag
from nltk.corpus import stopwords
from pymystem3 import Mystem
import re


download("stopwords")
download('averaged_perceptron_tagger_ru')
STOPWORDS = set(stopwords.words("russian"))
stemmer = Mystem()
```

```python
features_union = set()


def clear_diagnosis(diagnosis):
    diagnosis = diagnosis.lower()
    diagnosis = [word if word.isalpha() and not re.match('.*диагноз.*', word) \
                                        and not re.match(r'i{1,}', word)\
                                        and len(word) >= 5\
                 else '' for word in diagnosis.split(' ')]
    
    tokens = set(stemmer.lemmatize(' '.join(diagnosis)))
    
    tagged_tokens = pos_tag(tokens, lang = 'rus')
    
    tokens = [tagged[0] for tagged in tagged_tokens if tagged[1] == 'S']
    
    features_union.update(tokens)
    return ' '.join(tokens)

diagnosis_df['tokens'] = diagnosis_df.diagnosis.apply(clear_diagnosis)
len(features_union)
```

```python
diagnosis_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(diagnosis_df.tokens)
X.shape
```

```python
Y = diagnosis_df.has_insult
Y.shape
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import precision_score, precision_recall_curve


clf = SVC(C=6, kernel='poly', degree=3)

clf.fit(X_train, y_train)
```

```python
result = clf.predict(X_test)

precision_score(y_test, result)
```
